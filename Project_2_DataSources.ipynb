```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from functools import reduce
from sqlalchemy import create_engine
```

```python
#Files Used
#00 CountryList.xlsx
#01 countries of the world.csv
#02 Travel and Tourism direct contribution to GDP.xlsx
#03 self_employed_percent_of_employment.xlsx
#04 suicide-rates-by-country.xlsx
#05 API_ST.INT.ARVL_DS2_en_csv_v2_823405.csv
#06 share-population-female.csv
#07 Household Size and Composition.csv
```

```python
#File00 CountryList.xlsx
f0_path= "Data/CountryList.xlsx"
t0_read = pd.DataFrame(pd.read_excel(f0_path, sheet_name='Country'))
t0_read
```

```python
#File01 countries of the world.csv
f1_path= "Data/countries of the world.csv"
t1_read = pd.read_csv(f1_path, delimiter=',')
#Create dataframe and format
country_stats = pd.DataFrame({
                            'Country':t1_read['Country'],
                            #'Country_Code':"",
                            #'Year':0,
                            'travel_popdenisty_/sqmile':t1_read['Pop. Density (per sq. mi.)'],
                            'travel_coastline_%ratio':t1_read['Coastline (coast/area ratio)'],
                            'soc_gdp_percapita':t1_read['GDP ($ per capita)'],
                            'soc_literacy_%pop':t1_read['Literacy (%)'],
                            'eco_%agriculture':t1_read['Agriculture'],
                            'eco_%industry':t1_read['Industry'],
                            'eco_%service':t1_read['Service'],
                            'soc_birthrate_cbr':t1_read['Birthrate']
})

#country_stats['PopDenisty_per_mile_Ranked'] = country_stats['PopDenisty_per_mile'].rank(ascending=1)
#country_stats['Coast_Line_Ratio_Ranked'] = country_stats['Coast_Line_Ratio'].rank(ascending=0)
#country_stats['PhoneUse_per_1k_Ranked'] = country_stats['PhoneUse_per_1k'].rank(ascending=1)
#country_stats['Birthrate_Ranked'] = country_stats['Birthrate'].rank(ascending=0)

country_stats_df=pd.merge(t0_read, country_stats, on="Country")

country_stats_df
```

```python
#File02 Travel and Tourism direct contribution to GDP.xlsx
f2_path= "Data/Travel and Tourism direct contribution to GDP.xlsx"
intourism = pd.DataFrame(pd.read_excel(f2_path, sheet_name='Travel and Tourism direct contr'))

intourism_cln1=intourism['Indicator']=='Internal Travel and Tourism consumption'
intourism_cln2=intourism[intourism_cln1]
intourism_cln3=intourism_cln2['Subindicator Type']=="% growth"
intourism_cln4=intourism_cln2[intourism_cln3]

#Create dataframe and format
incountry_tourism = pd.DataFrame({
                            'Country':intourism_cln4['Country Name'],
                            #'Country_Code':intourism_cln4['Country ISO3'],
                            #'Year':2019,
                            'travel_incountry_%growth':intourism_cln4[2019]})
                            
#incountry_tourism['Percent_incountry_Travel_Growth_Ranked'] = incountry_tourism['Percent_incountry_Travel_Growth'].rank(ascending=0)
incountry_tourism = incountry_tourism.reset_index(drop=True)

incountry_tourism_df=pd.merge(t0_read, incountry_tourism, on="Country")

pd.set_option('display.max_rows', 200)
incountry_tourism_df
```

```python
#File03 self_employed_percent_of_employment.xlsx
f3_path= "Data/self_employed_percent_of_employment.xlsx"
self_emplyed = pd.DataFrame(pd.read_excel(f3_path, sheet_name='self_employed_percent_of_employ'))

#Create dataframe and format
self_emplyed = pd.DataFrame({
                            'Country':self_emplyed['country'],
                            #'Country_Code':"",
                            #'Year':2019,
                            'eco_%self_employed':self_emplyed[2019]})
                            
#self_emplyed['Percent_self_emplyed_Ranked'] = self_emplyed['Percent_self_emplyed'].rank(ascending=0)
self_emplyed = self_emplyed.reset_index(drop=True)

self_emplyed_df=pd.merge(t0_read, self_emplyed, on="Country")

pd.set_option('display.max_rows', 200)
self_emplyed_df
```

```python
#File04 suicide-rates-by-country.xlsx
f4_path= "Data/suicide-rates-by-country.xlsx"
suicide_by_country_df = pd.DataFrame(pd.read_excel(f4_path, sheet_name='suicide-rates-by-country'))

suicide_by_country_all=pd.DataFrame({
                            'Country':suicide_by_country_df['Entity'],
                            #'Country_Code':suicide_by_country_df['Code'],
                            #'Year':0,
                            'soc_suicide_per_100k':suicide_by_country_df['suicide rate (age-adjusted suicides per 100,000 people)']})

suicide_by_country = suicide_by_country_all.groupby(['Country']).agg({'soc_suicide_per_100k': 'mean'}).reset_index()

#suicide_by_country['suicide_per_100k_Ranked'] = suicide_by_country['suicide_per_100k'].rank(ascending=1)

suicide_by_country_df=pd.merge(t0_read, suicide_by_country, on="Country")

pd.set_option('display.max_rows', 200)

suicide_by_country_df
```

```python
#File05 API_ST.INT.ARVL_DS2_en_csv_v2_823405.csv (All arrival of travellers by Country by Year)
t5_path= "Data/API_ST.INT.ARVL_DS2_en_csv_v2_823405.csv"
t5_read = pd.read_csv(t5_path, delimiter=',',skiprows=4)
#Create dataframe and format
Country_tourism = pd.DataFrame({
                            'Country':t5_read['Country Name'],
                            'travel_no_of_arrivals':t5_read['2018']})
Country_tourism_df=pd.merge(t0_read, Country_tourism, on="Country")
#Country_tourism_df['No_of_Arrivals_Ranked'] = Country_tourism_df['No_of_Arrivals'].rank(ascending=0)
pd.set_option('display.max_rows', 200)

Country_tourism_df
```

```python
#File06 share-population-female.csv
f6_path= "Data/share-population-female.csv"
t6_read = pd.read_csv(f6_path, delimiter=',')

Per_of_FemalePop1=t6_read['Year']==2017
Per_of_FemalePop2=t6_read[Per_of_FemalePop1]

Per_of_FemalePop3 = Per_of_FemalePop2[Per_of_FemalePop2['Code'].notnull()]
Per_of_FemalePop4 = Per_of_FemalePop3[Per_of_FemalePop3['Entity']!="World"]

#Per_of_FemalePop3=Per_of_FemalePop2['Code']!=""
#Per_of_FemalePop4=t6_read[Per_of_FemalePop3]

Per_of_FemalePop = pd.DataFrame({
                            'Country':Per_of_FemalePop4['Entity'],
                            #'Country_Code':Per_of_FemalePop4['Code'],
                            #'Year':Per_of_FemalePop4['Year'],
                            'pop_%female':Per_of_FemalePop4['Population, female (% of total) (% of total)']
                                }).reset_index()
#Per_of_FemalePop['Per_of_FemalePop_Ranked'] = Per_of_FemalePop['Per_of_FemalePop'].rank(ascending=0)

Per_of_FemalePop_df=pd.merge(t0_read, Per_of_FemalePop, on="Country")

pd.set_option('display.max_rows', 200)
Per_of_FemalePop_df
```

```python
#File07 Household Size and Composition.csv
f7a_path= "Data/Household Size and Composition.csv"
t7a_read = pd.read_csv(f7a_path, delimiter=',',encoding ='latin1',skiprows=1)

colsa = [ 'Average household size (number of members)','Nuclear','Couple only']
t7a_read[colsa] = t7a_read[colsa].apply(pd.to_numeric, errors='coerce', axis=1)

t7b_read=t7a_read.groupby(['Country or area'])[colsa].mean().reset_index()
Household=pd.DataFrame({
                            'Country':t7b_read['Country or area'],
                            'pop_avghousehold_size':t7b_read['Average household size (number of members)'],
                            'pop_%householdtype_nuclear':t7b_read['Nuclear'],
                            'pop_%householdtype_nochildren':t7b_read['Couple only']
                            })

#Household['Household_size_Ranked'] = Household['Household_size'].rank(ascending=0)
#Household['Household_Nuclear_Ranked'] = Household['Household_Nuclear'].rank(ascending=0)

Household_df=pd.merge(t0_read, Household, on="Country")

pd.set_option('display.max_rows', 200)

Household_df
```

```python
#DataFrames
#1-country_stats
#2-incountry_tourism
#3-self_emplyed
#4-suicide_by_country
#5-Data Source World Development Indicators" (All arrival of travellers by Country by Year)
#6-Per_of_FemalePop
#7-Household

data_frames = [country_stats_df, incountry_tourism_df, self_emplyed_df,suicide_by_country_df\
               ,Country_tourism_df,Per_of_FemalePop_df,Household_df]

df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Country'],how='outer'), data_frames).fillna(0)

df_merged.drop('index', axis=1, inplace=True)  

pd.set_option('display.max_rows', 200)

df_merged
```

```python
happy_cols=['Country','travel_popdenisty_/sqmile','travel_incountry_%growth','travel_no_of_arrivals',\
            'soc_gdp_percapita','soc_literacy_%pop','soc_suicide_per_100k','soc_birthrate_cbr',\
            'eco_%agriculture','eco_%industry','eco_%service','eco_%self_employed',\
            'pop_%female','pop_avghousehold_size','pop_%householdtype_nuclear','pop_%householdtype_nochildren']

cols = ["Weight","Height","BootSize","SuitSize","Type"]
df_merged[happy_cols] = df_merged[happy_cols].replace(0, np.nan)

df_merged
```

```python
#08 Country List - Weighted Average
f8_path= "Data/CountryList.xlsx"
weighted_avg = pd.DataFrame(pd.read_excel(f8_path, sheet_name='Weight'))
weighted_avg
```

```python
#happy_indicator_df apply weighted average rank=Travel
happy_indicator_df=df_merged.copy()
tvl1= happy_indicator_df['travel_popdenisty_/sqmile'].rank(ascending=1)#Low is better
tvl2= happy_indicator_df['travel_coastline_%ratio'].rank(ascending=0) #High is better
tvl3= happy_indicator_df['travel_incountry_%growth'].rank(ascending=1)
tvl4= happy_indicator_df['travel_no_of_arrivals'].rank(ascending=0)

soc1= happy_indicator_df['soc_gdp_percapita'].rank(ascending=0)
soc2= happy_indicator_df['soc_literacy_%pop'].rank(ascending=0)
soc3= happy_indicator_df['soc_suicide_per_100k'].rank(ascending=1)
soc4= happy_indicator_df['soc_birthrate_cbr'].rank(ascending=0)

eco1= happy_indicator_df['eco_%agriculture'].rank(ascending=1)
eco2= happy_indicator_df['eco_%industry'].rank(ascending=1)
eco3= happy_indicator_df['eco_%service'].rank(ascending=0)
eco4= happy_indicator_df['eco_%self_employed'].rank(ascending=1)

pop1= happy_indicator_df['pop_%female'].rank(ascending=0)
pop2= happy_indicator_df['pop_avghousehold_size'].rank(ascending=0)
pop3= happy_indicator_df['pop_%householdtype_nuclear'].rank(ascending=0)
pop4= happy_indicator_df['pop_%householdtype_nochildren'].rank(ascending=1)

for index, row in happy_indicator_df.iterrows():
                happy_indicator_df['travel_rank'] = tvl1*[0.10]++tvl2*[0.20]++tvl3*[0.10]++tvl4*[0.60]
                happy_indicator_df['society_rank'] = soc1*[0.50]++soc2*[0.25]++soc3*[0.20]++soc4*[0.10]
                happy_indicator_df['economy_rank'] = eco1*[0.05]++eco2*[0.30]++eco3*[0.50]++eco4*[0.15]
                happy_indicator_df['pop_rank'] = pop1*[0.10]++pop2*[0.40]++pop3*[0.40]++pop4*[0.10]
                

all1= happy_indicator_df['travel_rank'].rank(ascending=1)
all2= happy_indicator_df['society_rank'].rank(ascending=1)
all3= happy_indicator_df['economy_rank'].rank(ascending=1)
all4= happy_indicator_df['pop_rank'].rank(ascending=1)
                
for index, row in happy_indicator_df.iterrows():
                happy_indicator_df['overall_rank'] = all1*[0.10]++all2*[0.35]++all3*[0.35]++all4*[0.15]                
            
happy_indicator_df
```

```python
happy_indicator=happy_indicator_df.copy()
happy_indicator['travel_rank']=happy_indicator_df['travel_rank'].rank(ascending=1)
happy_indicator['society_rank']=happy_indicator_df['society_rank'].rank(ascending=1)
happy_indicator['economy_rank']=happy_indicator_df['economy_rank'].rank(ascending=1)
happy_indicator['pop_rank']=happy_indicator_df['pop_rank'].rank(ascending=1)
happy_indicator['overall_rank']=happy_indicator_df['overall_rank'].rank(ascending=1)
happy_indicator
```

```python
happy_indicator=happy_indicator[['Country','travel_popdenisty_/sqmile','travel_coastline_%ratio','travel_incountry_%growth','travel_no_of_arrivals',\
                           'soc_gdp_percapita','soc_literacy_%pop','soc_suicide_per_100k','soc_birthrate_cbr',\
                           'eco_%agriculture','eco_%industry','eco_%service','eco_%self_employed',\
                           'pop_%female','pop_avghousehold_size','pop_%householdtype_nuclear','pop_%householdtype_nochildren',\
                           'travel_rank','society_rank','economy_rank','pop_rank','overall_rank']]

happy_indicator
happy_indicator.to_excel(r"Data/happy_indicator.xlsx",sheet_name='Indicator_By_Country', index =False, header=True)
```
